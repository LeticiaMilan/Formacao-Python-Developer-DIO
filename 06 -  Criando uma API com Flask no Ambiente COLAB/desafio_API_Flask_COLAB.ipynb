```python
from flask import Flask, jsonify
from flask_ngrok import run_with_ngrok
import pandas as pd

app = Flask(__name__)
run_with_ngrok(app)

table = {
    "Number": [1, 2, 3, 4, 5],
    "Name": ["Leticia", "Joao", "Larry", "Fernanda", "Henrique"],
    "Age": [21, 22, 27, 28, 29],
    "City": ["Batatais", "London", "San Francisco", "Toronto", "Paris"],
    "Country": ["Brazil", "UK", "USA", "Canada", "France"]
}

df = pd.DataFrame(table)

@app.route("/")
def home():
    return "<marquee><h3> Desafio DIO! </h3></marquee>"

@app.route("/input")
def input_data():
    return jsonify(df.to_json())

# MÃ©todo teste:
@app.route('/output', methods=['GET', 'POST'])
def output():
    return jsonify(df.to_dict())

if __name__ == "__main__":
    app.run()
```

```python
from flask import Flask
from flask_ngrok import run_with_ngrok
import pandas as pd

app = Flask(__name__)
run_with_ngrok(app)

table = {
    "Number": [1, 2, 3, 4, 5],
    "Name": ["Leticia", "Joao", "Larry", "Fernanda", "Henrique"],
    "Age": [21, 22, 27, 28, 29],
    "City": ["Batatais", "London", "San Francisco", "Toronto", "Paris"],
    "Country": ["Brazil", "UK", "USA", "Canada", "France"]
}

df = pd.DataFrame(table)

@app.route("/")
def home():
    return "<marquee><h3> Desafio DIO! </h3></marquee>"

@app.route("/input")
def input_data():
    return df.to_html()

@app.route('/output', methods=['GET', 'POST'])
def output():
    return df.to_html()

if __name__ == "__main__":
    app.run()
```

```python
from flask import Flask, jsonify, render_template
from flask_ngrok import run_with_ngrok

modelo = '/content/drive/MyDrive/Colab Notebooks'
app = Flask(__name__, template_folder=modelo)
run_with_ngrok(app)

table = {
    "Number": [1, 2, 3, 4, 5],
    "Name": ["Leticia", "Joao", "Larry", "Fernanda", "Henrique"],
    "Age": [21, 22, 27, 28, 29],
    "City": ["Batatais", "London", "San Francisco", "Toronto", "Paris"],
    "Country": ["Brazil", "UK", "USA", "Canada", "France"]
}

@app.route("/")
def home():
    return "<marquee><h3> Desafio DIO! </h3></marquee>"

@app.route("/input")
def input_data():
    return jsonify(table)

@app.route('/output')
def output():
    return render_template('representando_o_json_em_tabela.html', data=table)

if __name__ == "__main__":
    app.run()
```
